### This will read a Macro (a TPF/ALCS file for variable layout) text input and put it into excel file for easy documentation.

```python
import re
from pandas import DataFrame as df
f = open("data/Input_Macro.txt", "r")               # open macro file
ilist = []                                     # define input list
for x in f:                                    # for each line in macro file
    x = x.strip()                              # remove preceding and trailing spaces
    if x[0] == '*' or x[0] == '#':          # ignore lines starting with * and #
        continue           
    x = re.sub("&CG1" ,"  ", x)                # substitute &CG1 with spaces
    x = re.sub(" DS " ,"  ", x)                # substitite DS with spaces
    ilist.append(re.sub("\s\s+" , ",", x))     # substitute 2 or more spaces with one comma. This will convert
                                               # "RXMTSCAR        CL2            CARRIER" to "RXMTSCAR,CL2,CARRIER"
# We have created list of macro fields    
f.close()                                      # close the file
olist = [i.split(',') for i in ilist]          # create output list of lists by splitting the values with comma
olist = [trim[:4] for trim in olist]           # limit to 4 columns
olist = [trim for trim in olist if len(trim) > 1]  # remove lines that doesn't have required data
macrodf = df(data=olist,columns = ['Name','Length','Discription 1','Discription 2'])   # create a dataframe of macro fields

##new code to calculate actual length and displacement
######################################################

# create regex and dict for length format
len_syntax = re.compile('(\d*)?([ACDHFX])([L])?(\d*)?')   # Create a pattern for length syntax. Eg: 20CL7
dtypelen = {'A':4,'C':1,'D':8,'H':2,'F':4,'X':1}      # dict for length of each data type
length_list = macrodf['Length'].tolist()       # we need length column as a list

displacement = 0      
tot_arr = []                                   # list to store total length of each field
disp_arr = []                                  # list to store displacement of the field

# this loop will run for field's length, finds actual length syntax and calculates total and displacement.
for i in length_list:                          # for each length
    if 'WR10' in i or 'EQU' in i:              # if start of new LREC
        tot_arr.append(0)                      # length will be 0
        displacement = 0                       # displacement will start from 0. Very bad hard coding. Change it asap
        disp_arr.append(displacement)          
        continue
    oput = len_syntax.findall(i)               # find the pattern in length syntax
    oput = list(oput[0])                       # convert the tuple output to list: 20CL7 -> [('20','C','L','7')]
    oput = ['1' if v is '' else v for v in oput]    # CL4 will be [('','C','L','4')]. This shd change to [('1','C','L','4')]
    tot_len = int(oput[0])*int(oput[3])*dtypelen.get(oput[1])   # calculate total length for the field
    tot_arr.append(tot_len)                    
    disp_arr.append(displacement)
    displacement += tot_len                    # calculate displacement    
#####

macrodf['Actual Length'] = tot_arr             # add actual total and displacement lists to dataframe
macrodf['Displacement'] = disp_arr

macrodf.to_excel('data/macroexcel.xls', index = False)   # save it as excel
```
